```python
from src.dataset_wrapper import *
from src.networks import *
import torch
from tqdm import tqdm
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import pickle 

      
device = 'cuda:0'
net_params = {
    "relation_dim": 23,
    "object_dim": 8,
    "hidden_dim": 256,

}
dataset_params = {
    "device": device,
    "path": 'Data/cylinder_mass_9',
    "num_of_traj": 32000,
    "device": device,
    "data_split": (30000, 1000, 1000),
    "batch_size_train": 16,
    "batch_size_val": 32,
    'num_workers': 4,
    "num_of_batch_train": (0, 0), #(0, 10000),
    "num_of_batch_val":(0, 0), #(0, 500),
    'shuffle': True,
    "no_angle": True,
    'val_len': 40,
}
```

```python
Net = PropagationNetwork(**net_params).to(device)

dataset_params['Network'] = Net

dataset = DatasetWrapper(**dataset_params)
```

```python
metric1 = torch.nn.MSELoss(reduction='none')
metric2 = torch.nn.MSELoss()
def getMassErrorOverTime(ypred, ytarget):
    return metric1(ypred['obj_to_predict'], ytarget['obj_to_predict']).cpu().numpy()


with torch.set_grad_enabled(False):
    with torch.no_grad():
        predicted_br, ground_truth = dataset6.val_tester.test_br()
        mass_error_over_time6 = getMassErrorOverTime(
            predicted_br, ground_truth)
        del predicted_br, ground_truth
        predicted_br, ground_truth = dataset.val_tester.test_br()
        mass_error_over_time9 = getMassErrorOverTime(
            predicted_br, ground_truth)
        del predicted_br, ground_truth
        predicted_br, ground_truth = dataset12.val_tester.test_br(10)
        mass_error_over_time12 = getMassErrorOverTime(
            predicted_br, ground_truth)
        del predicted_br, ground_truth
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))
plt.plot(range(10,200),np.sqrt(mass_error_over_time9[:,10:,1:]).mean(axis=0).mean(axis=1).mean(axis=1),label='9 Objects',lw=4,ls=':',color='gray')
plt.plot(range(10,200),np.sqrt(mass_error_over_time6[:,10:,1:]).mean(axis=0).mean(axis=1).mean(axis=1),label='6 Objects',lw=4,ls='--',color='gray')
plt.plot(range(10,200),np.sqrt(mass_error_over_time12[:,10:,1:]).mean(axis=0).mean(axis=1).mean(axis=1),label='12 Objects',lw=4,color='gray')
plt.legend(fontsize=18)
plt.xlabel('Number of Observed Timesteps',fontsize=20)
plt.ylabel('Mass Error(kg)',fontsize=20)
plt.title("Mass Error Over Time",fontsize=20)
plt.grid()
plt.tick_params(labelsize=16)
plt.savefig("me_over_time.png",bbox_inches='tight')
```

```python
dataset2_params = {
    "device": device,
    "path": 'Data/cylinder_mass_test_2',
    "num_of_traj": 1000,
    "device": device,
    "data_split": (0, 0, 1000),
    "batch_size_train": 16,
    "batch_size_val": 32,
    'num_workers': 4,
    "num_of_batch_train": (0, 0), #(0, 10000),
    "num_of_batch_val":(0, 0), #(0, 500),
    'shuffle': True,
    "no_angle": True,
    "scaler": dataset.scaler,
    'val_len': 40,
}
dataset2_params['Network'] = Net

dataset2 = DatasetWrapper(**dataset2_params)

dataset3_params = {
    "device": device,
    "path": 'Data/cylinder_mass_test_3',
    "num_of_traj": 1000,
    "device": device,
    "data_split": (0, 0, 1000),
    "batch_size_train": 16,
    "batch_size_val": 32,
    'num_workers': 4,
    "num_of_batch_train": (0, 0), #(0, 10000),
    "num_of_batch_val":(0, 0), #(0, 500),
    'shuffle': True,
    "no_angle": True,
    "scaler": dataset.scaler,
    'val_len': 40,
}

dataset3_params['Network'] = Net

dataset3 = DatasetWrapper(**dataset3_params)

```

```python
metric1 = torch.nn.MSELoss(reduction='none')
metric2 = torch.nn.MSELoss()
def getMassErrorOverTime(ypred, ytarget):
    return metric1(ypred['obj_to_predict'], ytarget['obj_to_predict']).cpu().numpy()

with torch.set_grad_enabled(False):
    with torch.no_grad():
        predicted_br2, ground_truth2 = dataset2.val_tester.test_br(10,74)
        mass_error_over_time2 = getMassErrorOverTime(
            predicted_br2, ground_truth2)
        del predicted_br2, ground_truth2
me_over_time2 = mass_error_over_time2.mean(axis=(2,3))
with torch.set_grad_enabled(False):
    with torch.no_grad():
        predicted_br3, ground_truth3 = dataset3.val_tester.test_br(10,74)
        mass_error_over_time3 = getMassErrorOverTime(
            predicted_br3, ground_truth3)
        del predicted_br3, ground_truth3
me_over_time3 = mass_error_over_time3.mean(axis=(2,3))
```

```python
fig,ax = plt.subplots(1,2,figsize=(15,5))
num_of_objects=['2 ','3 ' ]
for ind,error_pos in enumerate([me_over_time2,me_over_time3]):
    ax[ind].set_title(num_of_objects[ind]+'Objects',fontsize=20)
    bins_ = [-1.5,-0.5,0.5,1.5,2.5,3.5]
    xticks =[-1,0,1,2,3,4]
    xtick_labels = ['< 1 kg','1-2 kg','2-4 kg','4-8 kg','>8 kg']

    data = np.log2(np.sqrt(error_pos[:,73]))
    data[data<-1.]=-1
    ax[ind].hist(data,bins=bins_,color='gray',rwidth=0.8)
    plt.sca(ax[ind])
    plt.yticks(fontsize=16)
    plt.xticks(xticks,xtick_labels,rotation=60,fontsize=16)
    fig.suptitle("Mass Errors (kg)",fontsize=20)
    ax[ind].set_ylim([0,350])
ax[1].get_yaxis().set_ticks([])

ax[0].set_axisbelow(True)
ax[0].grid()
ax[1].set_axisbelow(True)
ax[1].grid()

plt.savefig("me_2_3_new.png",bbox_inches='tight')
```
